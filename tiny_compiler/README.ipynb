### Virtual Machine Design

1. Stack: from 0 to 127 words

2. Registers: `PC`, `SP` & `FP`

3. Instructions:
    - `add`: stack[SP - 1] = stack[SP - 1] + stack[SP]; SP -= 1;
    - `sub`: stack[SP - 1] = stack[SP - 1] - stack[SP]; SP -= 1;
    - `mul`: stack[SP - 1] = stack[SP - 1] * stack[SP]; SP -= 1;
    - `div`: stack[SP - 1] = stack[SP - 1] / stack[SP]; SP -= 1;
    - `mod`: stack[SP - 1] = stack[SP - 1] % stack[SP]; SP -= 1;
    - `loadi`: SP += 1; stack[SP] = imm;
    - `load`: stack[SP + 1] = stack[FP + imm]; SP += 1;
    - `store`: stack[FP + imm] = stack[SP]; SP -= 1;
    - `syscall`: scan & print & terminate
    - `jmp`: PC = imm;
    - `rjmpz`: if stack[SP] == 0: PC += imm; else: PC += 1; SP -= 1;
    - `rjmp`: PC += imm;
    - `save`:
    - `call`:
    - `ret`:
    - `push`: SP += imm; PC += 1;
    - `pop`: SP -= imm; PC += 1;

### Calling Convention

1. Caller: 
    - saves PC & FP
    - pushes arguments
    - sets new PC & FP

2. Callee:
    - pushes return values
    - recovers PC & FP

To compile an arbitrary source and run:

```python
from topcg import TopCodeGenerator
from vm import VirtualMachine

source = """
// write your source code here
func main(void) -> 0 {
    var a = 1;
    var b = 2;
    a = a + b;
    return (void);
}
"""

vm = VirtualMachine()
cg = TopCodeGenerator()
try:
    code = cg.compile(source)
    vm.execute(code)
except Exception as e:
    print(e)
```
